```python
#operators and types
# Q6 - What is the result and type of each?
print(10 / 3)   # <class'float'>
print(10 // 3)  # <class'int'>
print(10 % 3)   # <class'int'>
print(2 ** 8)   # <class'int'>
```

```python
# Q7 - A cinema has 15 rows with 28 seats each
# Define rows = 15
# Define seats_per_row = 28
# Calculate total_seats
# Calculate remaining_seats if 247 people are seated
# Calculate percentage of seats filled (round to 2 decimal places)
# Q7 - Cinema seats
rows = 15
seats_per_row = 28
total_seats = rows * seats_per_row
print(f"Total seats: {total_seats}")

seated_people = 247
remaining_seats = total_seats - seated_people
print(f"Remaining seats: {remaining_seats}")

per_of_seats_filled = round((seated_people / total_seats) * 100, 2)
print(f"Percentage filled: {per_of_seats_filled}%")
```

```python
# Q8 - Temperature converter
# temp_celsius = 37
# Convert to Fahrenheit using formula: (celsius * 9/5) + 32
# Print the result
temp_celsius = 37
fahrenheit = (temp_celsius * 9 / 5) + 32
print(fahrenheit)
```

```python
# Q9 - A pizza is cut into 8 slices
# 3 people want equal shares
# How many full slices does each person get? (use //)
# How many slices are left over? (use %)
slices = 8
people = 3
each_share = slices // people
print(each_share)
remained_slices = slices % people
print(remained_slices)
```

```python
# Q10 - Define these variables
# principal = 10000 (money invested)
# rate = 0.05 (5% interest)
# time = 3 (years)
# Calculate simple interest = principal * rate * time
# Calculate total amount = principal + simple_interest
principal = 10000
rate = 0.05
time = 3
simple_interest = principal * rate * time
print(simple_interest)
total_amount = principal + simple_interest
print(total_amount)
```
