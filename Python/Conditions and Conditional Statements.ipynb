```python
# everyone earning less than 12,000 pays 25 % in taxes, and anyone earning12,000 0r more pays 30% , calculate how much tax is owned
def get_taxes(earnings):
    if earnings < 12000:
        tax_owned = .25 * earnings
    else:
        tax_owned = .30 * earnings
    return tax_owned
```

```python
ana_taxes = get_taxes(9000)
bob_taxes = get_taxes(15000)
```

```python
print(ana_taxes)
print(bob_taxes)
```

```python
# accepts a number as input and add threee if the input is less than 10, and otherwise adds eight
def add_three_or_eight(number):
    if number < 10:
        result = number + 3
    else:
        result = number + 8
    return result
```

```python
new_number = add_three_or_eight(9)
print(new_number)
```
