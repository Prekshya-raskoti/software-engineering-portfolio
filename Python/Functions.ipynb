# Functions
Organize our code and avoid redundancy

```python
# function accepting any number, add 3 to it and return the result
def add_three(input_var):
    output_var = input_var + 3
    return output_var
    
```

```python
new_number = add_three(100)
print(new_number)
```

```python
# calculating weekly paycheck after taxes.
#(they are in 12 % tax bracket meaning 12 % of their salary is taken for taxes, and they only take home 88%)
# they are paid hourly, at a rate of $15/hour.
def get_pay(num_hours):
    pay_pretax = num_hours * 15
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
    
```

```python
pay_fulltime = get_pay(40)
print(pay_fulltime)
```

```python
pay_parttime = get_pay(32)
print(pay_parttime)
```

```python
# Variables defined inside the function body cannot be accessed outside the function
# Variables defined outside the function body can be accessed outside the function (global scope)
```

# Functions with Multiple Arguments

```python
 def get_pay_with_more_inputs (num_hours,hourly_wages,tax_bracket):
     pay_pretax = num_hours * hourly_wages
     pay_aftertax = pay_pretax * (1- tax_bracket)
     return pay_aftertax
     
```

```python
fulltime_pay_aftertax = get_pay_with_more_inputs(40,20,.30)
print(fulltime_pay_aftertax)
```
