# Data Types
- Whenever we create a variable to python, it has a value with a corresponding data type.


```python
x = 4
print(x)
print(type(x))
```

```python
nearly_pi = 3.1467884363748
print(nearly_pi)
print(type(nearly_pi))
```

```python
# round() function- it lets to round a number to a specified number of decimal places
# rounding to 5 decimal places
round_pi = round(nearly_pi,5)
print(round_pi)
print(type(round_pi))
```

```python
# for instance, 1. , 1.0 , 1.00 , etc.are recognized as a float
y = 1.
print(y)
print(type(y))
```

# Booleans represent one of two values: True or False

```python
z_one = True
print(z_one)
print(type(z_one))
```

```python
z_two = False
print(z_two)
print(type(z_two))
```

```python
z_three = (1 < 2)
print(z_three)
print(type(z_three))
```

```python
z_four = (5 < 3)
print(z_four)
print(type(z_four))
```

```python
z_five = not z_four
print(z_five)
print(type(z_five))
```

## Strings

```python
W = "Hello python"
print(W)
print(type(W))
```

```python
print(len(W))
```

```python
## One special type of string is the empty string, which has length zero.
shortest_string = ""
print(type(shortest_string))
print(len(shortest_string))
```

```python
# if we put a number in quotation marks,it has a string data type
my_number = "1.234567"
print(my_number)
print(type(my_number))
```

```python
new_string = "abc" + "def"
print(new_string)
print(type(new_string))
```

```python
newest_string = "abc" * 3
print(newest_string)
print(type(newest_string))
```
