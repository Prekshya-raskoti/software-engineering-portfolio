```python
flowers_list = ["Pink primrose"," sweet pea","tiger lily"]
print(type(flowers_list))
print(flowers_list)
```

```python
print(len(flowers_list))
```

## Python uses zero based indexing

```python
print("first entry:", flowers_list[0])
print("second entry:", flowers_list[1])
```

## Slicing 
- pulling a segment of a list

```python
print("first two entries:", flowers_list[:2])
```

```python
print("final two entries:", flowers_list[-2:])
```

## Removing items

```python
flowers_list.remove("tiger lily")
print(flowers_list)
```

```python
# removing using index 
numbers = [10,20,30]
numbers.pop(1)
print(numbers)
```

```python
colors = ["red","green","yellow"]
del colors[0]
print(colors)
```

```python
my_set = {1,2,3}
my_set.remove(2)
print(my_set)
              
```

```python
student = {"name": "Prekshya", "age":22}
student.pop("age")
print(student)
```

```python
# adding items
flowers_list.append("lotus")
print(flowers_list)
```

## Lists are not just for strings - any datatypes including booleans, integers and floats

```python
hardcover_sales = [139,123,345,678,890,987]
print("lenght of the list:",len(hardcover_sales))
print("entry at index 2:", hardcover_sales[2])
```

```python
print("Maximum:",max(hardcover_sales))
print("Minimum:",min(hardcover_sales))
```

```python
print("sum:", sum(hardcover_sales))
```

```python
# average book sold in first five days
print("Average books sold in first five days:",sum(hardcover_sales[:5])/5)
```
