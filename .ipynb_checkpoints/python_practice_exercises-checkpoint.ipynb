```python
#variables and maths
# Q1 - Store your name and age in variables and print them
name = "Prekshya Raskoti"
age = 23
print("My name is", name,"and I am",age, "years old.") 
print(f"My name is {name} and i am {age} years old.")
```

```python
# Q2 - There are 7 days in a week and 52 weeks in a year
# Define days_per_week = 7
# Define weeks_per_year = 52
# Calculate days_per_year using those two variables
days_per_week = 7
weeks_per_year = 52
days_per_year = days_per_week * weeks_per_year
print(days_per_year) 
```

```python
# Q3 - A shop sells 35 items per hour
# Define items_per_hour = 35
# Calculate items_per_day (shop open 8 hours)
# Calculate items_per_week
items_per_hour = 35
items_per_day = items_per_hour * 8
print(items_per_day)
items_per_week = items_per_hour * 8 * 7
print(items_per_week)
```

```python
# Q4 - Define a variable speed_kmh = 60
# Calculate how far a car travels in 3.5 hours
# distance = speed * time
speed_kmh = 60
time = 3.5 
distance = speed_kmh * time
print(distance)
```

```python
# Q5 - A student reads 25 pages per day
# Calculate how many pages in a week
# Calculate how many pages in a month (30 days)
# Calculate how many pages in a year
pages_per_day = 25
pages_per_week = pages_per_day * 7
print(pages_per_week)
pages_per_month = pages_per_day * 30
print(pages_per_month)
pages_per_year = pages_per_day * 365
print(pages_per_year)
      
```
